```python
# please call the 'set_browser_url' function first

def menu_help_acc():
    # click menu
    set_browser_url.driver.find_element_by_css_selector("div.sticky-outer-wrapper div.sticky-inner-wrapper header.header.overlay-header \
                                                div.user-wrapper div.account button.account-link \
                                                span.header__account__toggle__arrow > svg:nth-child(1)").click()

    time.sleep(3)

    # click help center
    set_browser_url.driver.find_element_by_css_selector("body.country-bd.home.not-mobile.show-banner.show-rider-banner:nth-child(2) \
                                        div.top-section:nth-child(1) header.header.header--white:nth-child(2) div.user-wrapper \
                                        div.account.header__account div.header__account__logged div.header__account__dropdown.\
                                        dropdown.open ul.dropdown-menu.dropdown-menu-sm.arrow li:nth-child(7) > \
                                        a.header_account__dropdown__link.js-self-service").click()

    time.sleep(5)
    # click my account
    set_browser_url.driver.find_element_by_id('account').click()
```

**define a function to go with yes -> some -> submit -> keep navigation**

```python
# please call the 'menu_help_acc' function first

def yes_to_navigate():
    # click yes
    buttons = set_browser_url.driver.find_elements_by_tag_name('button')
    for button in buttons:
        btn = button.find_element_by_class_name('PxyKF button__content').text
        if 'Yes' in btn:
            button.click()

    time.sleep(2)
    # click 'something else' button
    div = set_browser_url.driver.find_element_by_class_name('B2Npn')
    buttons = div.find_elements_by_tag_name('button')
    for button in buttons:
        btn = button.find_element_by_class_name('FvM5h button__text').text
        if 'Something' in btn:
            button.click()

    time.sleep(2)
    # click submit
    set_browser_url.driver.find_element_by_class_name('uEm8v u3D3O BVKYX').click()

    thank_you = set_browser_url.driver.find_element_by_class_name('gfuqT').text

    time.sleep(2)
    # click keep navigation
    nv_btn = set_browser_url.driver.find_element_by_tag_name('button')
    if 'Keep' in nv_btn.text:
        nv_btn.click()

    assert 'Thank' in thank_you, 'Thank you message not visible'
    print('Thank you message visible')
```
